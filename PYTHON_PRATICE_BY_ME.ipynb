```python
# Define a class (blueprint for creating objects)
class Computer:  
    # Constructor method (runs automatically when an object is created)
    def __init__(self, brand):
        # self refers to the current object
        # brand parameter receives the value passed during object creation
        # self.brand creates an attribute called brand for the object
        self.brand = brand
# Creating first object of the Computer class
# "Lenovo" is passed to the constructor
c1 = Computer("Lenovo")
# Creating second object
# "Dell" is passed to the constructor
c2 = Computer("Dell")
# Accessing the brand attribute using the object
print(c1.brand)   # prints the brand of object c1
# Accessing the brand attribute of the second object
print(c2.brand)   # prints the brand of object c2
```

```python
class student:
    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2
    def __add__(self,other): #here over loading method is used because # Overloads the + operator for student objects
        m1 = self.m1 + other.m1
        m2 = self.m2 + other.m2
        # Create a new student object with summed marks
        s3 = student(m1,m2)
        return s3

    def __gt__(self,other):   # Overloads > operator to compare two student objects
        r1 = self.m1 + self.m2
        r2 = other.m1 + other.m2
        if r1 > r2 :
            return True
        else:
            return False

            
    def __str__(self):# Overloads print() so object displays nicely
        # Returns marks as a formatted string instead of object address
        return '{} {}' .format(self.m1,self.m2)     
    
            
s1 = student(58,69)
s2 = student(60,65)
# Add the two objects (calls __add__ method)
s3 = s1 + s2
print(s3.m1)

# Compare two students (calls __gt__)
if s1 > s2:
    print('s1 wins')
else:
    print('s2 wins')

print(s1)
print(s2)
```

```python
# linear search

a= [1,6,8,4,3,10]
n = int(input("enter the number")) # entering the input from user
def search(a,n):  # we defined the fuction for search
     for i in range(len(a)):
         if a[i] == n:
             return i
            
     return -1
result = search(a, n)

if result != -1:
    print(f"{n} found at index {result}")
else:
    print(f"{n} not found")
         
        
    
        
    
```

```python
# binary search

a= [1,6,8,4,3,10]
n = int(input("enter the number")) # entering the input from user
low = 0
upper =len(a)-1
def binary_search(a,n):# we defined the fuction for search
    global low,upper   # here gloabl keyword is used
    while low <= upper:
        mid = (low + upper) // 2

        if a[mid] == n:
            return mid
        elif a[mid] < n:
            low =mid+1
        elif a[mid]> n:
            upper =mid-1
    return -1
result = binary_search(a,n)
if result != -1:  # id -1 is not found means the index should wiil come from 0 not -1 
    print(f"{n} found at index {result}")    '''   f"text {variable} text"  '''' # formatted string is used  
else:
    print(f"{n} not found")
            
        
        
    
```

```python
# bubble sort
a= [1,6,8,4,3,10]

def bubble_sort(a):
    for i in range(len(a)-1,0,-1):
        for j in range(i):
            if a[j] > a[j+1]:
                a[j],a[j+1] =a[j+1],a[j] # we changing the index i.e swaping 
bubble_sort(a)
print(a)


            
```

```python
#  selection sort
a= [1,6,8,4,3,10]

def selection_sort(a):
    for i in range(len(a)-1):
       min_index = i
       for j in range(i + 1, len(a)):
            if a[j] < a[min_index]:
                min_index = j
       a[i], a[min_index] = a[min_index], a[i]

selection_sort(a)
print(a)
```
